```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import numpy as np
import matplotlib as mpl
```

```python
import datetime
from datetime import date
```

```python
mpl.rcParams['figure.figsize'] = (12,6)
mpl.rcParams['font.size'] = 18
```

```python
dados_internacoes= pd.read_csv("https://raw.githubusercontent.com/seade-R/dados-covid-sp/master/data/plano_sp_leitos_internacoes_serie_nova_variacao_semanal.csv", sep=';', thousands=".", decimal=",")
```

```python
dados_internacoes.head()
```

```python
dados_internacoes['nome_drs'].unique()
```

```python
 dados_internacoes['data'] = pd.to_datetime(dados_internacoes['datahora'],format='%Y-%m-%d')
```

```python
dados_internacoes['ano']= dados_internacoes['data'].dt.year
```

```python
internacoes_grande_sp= dados_internacoes.query('nome_drs == "DRS 01 Grande São Paulo" ')
```

```python
internacoes_grande_sp = internacoes_grande_sp.query('ano == 2021')
```

```python
internacoes_grande_sp.head()
```

```python
internacoes_grande_sp.plot(x= 'data', y= 'internacoes_ultimo_dia', figsize=(12,8))
internacoes_grande_sp.plot(x='data', y= 'pacientes_uti_mm7d', figsize=(12,8))
```

```python
font_t = {'family': 'sans-serif',
        'color': '#777D62',
        'weight': 'bold'
         }
```

```python
sns.lineplot(x='data',y='pacientes_uti_mm7d',data=internacoes_grande_sp, color="#FF9962").set_title('Média móvel de pacientes em leitos de UTI na Grande SP', pad = 30, fontdict=font_t)
plt.xlabel('Data', fontdict= font_t)
plt.ylabel('Pacientes em UTI COVID-19', fontdict= font_t, labelpad= 20)
plt.tick_params(labelcolor= '#777D62')
plt.ylim(0,8000)
plt.show()
```

```python
dados_casos_sp = pd.read_csv('https://raw.githubusercontent.com/seade-R/dados-covid-sp/master/data/dados_covid_sp.csv', sep=';', thousands=".", decimal=",")
```

```python
dados_casos_sp
```

```python
 dados_casos_sp['data'] = pd.to_datetime(dados_casos_sp['datahora'],format='%Y-%m-%d')
```

```python
dados_casos_sp['ano']= dados_casos_sp['data'].dt.year
```

```python
lista_cidades= ['ARUJÁ', 'BARUERI', 'BIRITIBA-MIRIM', 'CAIEIRAS', 'CAJAMAR', 'CARAPICUÍBA', 'COTIA', 'DIADEMA', 'EMBU', 'EMBU-GUAÇU', 'FERRAZ DE VASCONCELOS'
'FRANCISCO MORATO' , 'FRANCO DA ROCHA', 'GUARAREMA', 'GUARULHOS', 'ITAPECERICA DA SERRA', 'ITAPEVI', 'ITAQUAQUECETUBA', 'JANDIRA', 'JUQUITIBA', 'MAIRIPORÃ', 
'MAUÁ', 'MOGI DAS CRUZES', 'OSASCO', 'PIRAPORA DO BOM JESUS', 'POÁ', 'RIBEIRÃO PIRES', 'RIO GRANDE DA SERRA', 'SALESÓPOLIS', 'SANTA ISABEL', 'SANTANA DE PARNAÍBA', 
'SANTO ANDRÉ', 'SÃO BERNARDO DO CAMPO', 'SÃO CAETANO DO SUL', 'SÃO LOURENÇO DA SERRA', 'SÃO PAULO', 'SUZANO', 'TABOÃO DA SERRA', 'VARGEM GRANDE PAULISTA']
```

```python
lista_drs_grande_sp=[]
```

```python
for x in lista_cidades:
     cidade = x.lower().capitalize()
     lista_drs_grande_sp.append(cidade)
```

```python
lista_drs_grande_sp
```

```python
casos_drs_grande_sp = dados_casos_sp.query('nome_munic in @lista_drs_grande_sp')
```

```python
casos_drs_grande_sp = casos_drs_grande_sp.query('ano == 2021')
```

```python
casos_drs_grande_sp.head()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data = casos_drs_grande_sp, x= 'data', y='casos_novos')
```

```python
casos_por_dia = casos_drs_grande_sp.groupby("data").sum()
casos_por_dia= casos_por_dia.reset_index()
```

```python
casos_media_movel = casos_por_dia[['data','casos_mm7d', 'casos_novos']]
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x = 'data', y = 'casos_mm7d', data = casos_media_movel)
```

```python
sns.lineplot(x='data',y='casos_mm7d',data=casos_media_movel, color="#FF9962").set_title('Média semanal de Casos novos de COVID-19 na Grande SP em 2021', pad = 30, fontdict=font_t)
plt.ylabel('Casos', fontdict= font_t)
plt.xlabel('Data', fontdict= font_t)
plt.ylim(0,1600)
plt.tick_params(labelcolor= '#777D62')
```
